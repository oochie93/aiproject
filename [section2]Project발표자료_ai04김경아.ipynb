-마키아벨리즘이란? 

-마키아벨리즘 성향 높은 사람들은 다른 사람들과 소통하는 데 있어 보다 계산적이고 신중하게 접근하는 경향이 있고, 반면에 성향 낮은 사람들은 보다 타인과의 소통에 있어서 보다 개인적이며 감정이 이입된 접근을 하고자 한다. 이 사람들은 성선설을 믿으며, 이로 인해 타인을 믿는 경향이 강하고 보다 정직한 태도를 보입니다.

-심리테스트는 총27문항으로 이루어져 있어 간편하게 테스트할 수 있습니다.

-데이터의 출처: 데이콘 대회에서 진행되었던 데이터셋을 일부 활용하였습니다. 



**<데이터 선정 이유>**

 사람마다 각각의 성격이나 가치관, 다양한 백그라운드 등을 분석하고 특정 패턴을 찾아냄으로써, 사회적으로 중요한 사안을 예측해볼 수 있다는 점에 흥미를 느꼈기 때문에 본 데이터를 고르게 되었습니다.

### 특성 소개와 타겟설정


[QaA ~ QtA] : 20문항의 질문 내용 / 1=Disagree, 2=Slightly disagree, 3=Neutral, 4=Slightly agree, 5=Agree.     
Qa : 미공개            
Qb : 범죄자들과 아닌 사람들의 가장 큰 차이점은 그들이 잡힐만큼 어리석다는 것이다.            
Qc : 다른 사람을 완전히 신뢰하는 사람은 화를 자초한다.           
Qd : 미공개           
Qe : P.T. Barnum의 매분마다 멍청한 사람이 태어난다라는 말은 틀렸다
Qf : 다른 사람에게 거짓말을 할 이유가 없다             
Qg : 미공개                
Qh : 대부분의 사람들은 재산을 잃는 것보다 부모의 죽음을 더 쉽게 잊는다.        
Qi : 미공개        
Qj : 모든 사람들은 잔인한 면을 갖고 있어서 기회가 주어지면 그 본성이 드러난다고 생각하는 것이 가장 안전하다.            
Qk : 대체로 중요하고 부정직한 것보다 겸손하고 정직한 것이 낫다.            
Ql : 미공개             
Qm : 이곳 저곳 절차를 생략하지 않고서는 일을 추진하는 것이 어렵다.      
Qn : 미공개            
Qo : 사람들을 다루는 가장 좋은 방법은 그들이 듣고 싶은 것을 말하는 것이다.    
Qp : 미공개            
Qq : 대부분의 사람들은 기본적으로 착하고 친절하다.          
Qr : 도덕적으로 옳다고 확신 할 때만 행동을 취해야 한다.          
Qs : 중요한 사람들에게는 아첨하는 것이 현명하다.            
Qt : 미공개           


[QaE ~ QtE] : 질문을 답할 때까지의 시간    
age_group : 연령    
education : 교육 수준    
1=Less than high school, 2=High school, 3=University degree, 4=Graduate degree, 0=무응답

engnat : 모국어가 영어
1=Yes, 2=No, 0=무응답

familysize : 형제자매 수

gender : 성별
Male, Female

hand : 필기하는 손
1=Right, 2=Left, 3=Both, 0=무응답

married : 혼인 상태
1=Never married, 2=Currently married, 3=Previously married, 0=Other

race : 인종
Asian, Arab, Black, Indigenous Australian, Native American, White, Other

religion : 종교

urban : 유년기의 거주 구역
1=Rural (country side), 2=Suburban, 3=Urban (town, city), 0=무응답

[wr_01 ~ wr_13] : 13개의 실존단어의 정의을 안다
1=Yes, 0=No
[wf_01 ~ wf_3] : 3개의 허구단어의 정의를 안다
1=Yes, 0=No

voted (타겟): 지난 해 국가 선거 투표 여부
1=Yes, 2=No

[tp01 ~ tp10] : 자기성격평가 / 1(강한 긍정) ~ 7(강한 부정)    
tp01 : 외향적이고 열정적이다.    
tp02 : 비판적이고 논쟁적이다.    
tp03 : 신뢰할 수 있고 자제력이 있다.    
tp04 : 불안하고 쉽게 화를 낸다.    
tp05 : 새로운 경험에 개방적이며 복잡하다.    
tp06 : 내성적이고 조용하다    
tp07 : 동정적이고 따뜻하다    
tp08 : 무질서하고 부주의하다   
tp09 : 차분하고 정서적으로 안정적이다    
tp10 : 전통적이고 창조적이지 않다

```
def pre(df):
    age_replace = {'10s':'10', '20s':'20', '30s':'30', '40s':'40', '50s':'50', '60s':'60', '+70s':'70'} 
    df['age_group'] = df['age_group'].map(age_replace).astype(int)  #age_group: 's' 제거 후 정수변환
    df['gender'] = df['gender'].apply(lambda x: 0 if 'Female' in x else 1)  #gender: Female = 0 , Male = 1로 변환
    df['education'] = df['education'].replace(0,df['education'].mode()[0])
    df['engnat'] = df['engnat'].replace(0,df['engnat'].mode()[0])
    df['hand'] =  df['hand'].replace(0, df['hand'].mode()[0])
    df['married'] = df['married'].replace(0,df['married'].mode()[0])
    df['urban'] = df['urban'].replace(0,df['urban'].mode()[0])
    for i in range(1,9):
       df['tp0'+str(i)] = df['tp0'+str(i)].replace(7, df['tp0'+str(i)].mode()[0])
    df['tp10'] = df['tp10'].replace(7, df['tp10'].mode()[0])
    return df
df = pre(df)

#필요한 컬럼 추가/ 불필요한 것 제거
df['wf'] = df['wf_01'] + df['wf_02'] + df['wf_03']
df['wr'] = df['wr_01'] + df['wr_02'] + df['wr_03'] +df['wr_04'] +df['wr_05'] + df['wr_06'] +df['wr_07'] + df['wr_08'] + df['wr_09'] + df['wr_10'] + df['wr_11'] + df['wr_12'] + df['wr_13'] 
df = df.drop(['wf_01','wf_02','wf_03','wr_01','wr_02','wr_03','wr_04','wr_05','wr_06','wr_07','wr_08','wr_09','wr_10','wr_11','wr_12','wr_13'],axis=1)
df['M_score'] = df.loc[:, 'QaA':'QtA':2].sum(axis=1) #QaA~QtA까지의 합
df
```

#  4. 머신러닝 방식 적용 및 교차검증

기준모델

```
#baseline   타겟변수인 voted의 클래스 불균형은 없다고 볼 수 있다.
target = df['voted']
target.value_counts(normalize=True)   
```

```
target = 'voted'
features = df.drop([target],axis=1).columns
train, test = train_test_split(df, train_size = 0.8, test_size = 0.2, random_state = 2)
X_train = train[features]
y_train = train[target]
X_test = test[features]
y_test = test[target]
```

첫번째 모델링 Randomforest 

```
pipe = make_pipeline(
    OrdinalEncoder(cols=['race','religion'])
    RandomForestClassifier(n_estimators=100, random_state=2, n_jobs=-1)
)
pipe.fit(X_train, y_train)

print('훈련 정확도: ', pipe.score(X_train,y_train))
print('테스트 정확도: ',pipe.score(X_test,y_test))

# 예측 및 평가 
pred_train = pipe.predict(X_train)
pred_test = pipe.predict(X_test)
train_score = roc_auc_score(y_train, pred_train)
test_score =roc_auc_score(y_test, pred_test)
print('훈련 Roc_AUC_score:', train_score)
print('테스트 Roc_AUC_score:', test_score)
```

두번째 모델링 XGBoost

```
pipe1 = make_pipeline(
    OrdinalEncoder(),
    XGBClassifier(n_estimators=200
                  , random_state=2
                  , n_jobs=-1
                  , max_depth=7
                  , learning_rate=0.2
                 )
)

pipe1.fit(X_train, y_train);
```

```
pred_train = pipe1.predict(X_train)
pred_test = pipe1.predict(X_test)

train_score = roc_auc_score(y_train, pred_train)
test_score =roc_auc_score(y_test, pred_test)
test_score 

print('훈련 정확도: ', pipe1.score(X_train,y_train))
print('테스트 정확도: ',pipe1.score(X_test,y_test))
print('훈련 Roc_AUC_score:', train_score)
print('테스트 Roc_AUC_score:', test_score)
```

```
pipe2 = make_pipeline(
    OrdinalEncoder(),
    XGBClassifier(
        random_state = 2,
        n_jobs = -1
    )
)

params = {
    'xgbclassifier__n_estimators' : [30,50],
    'xgbclassifier__learning_rate' : [0.1, 0.2, 0.3],
    'xgbclassifier__max_depth' : [5, 7, 10, 15]

}
clf = GridSearchCV(
    pipe2,
    param_grid=params,
    cv = 3,
    scoring='roc_auc',
    n_jobs = -1,
    verbose = 1
    )
'''
clf = RandomizedSearchCV(
    pipe2, 
    param_distributions=params, 
    n_iter=50, 
    cv=3,
    scoring='f1',
    verbose=1,
    n_jobs=-1
)
'''
clf.fit(X_train, y_train);
```

```
xgb_model = clf.best_estimator_
print('최적 하이퍼파라미터: ', clf.best_params_)
print('roc_auc score: ', -clf.best_score_)
```

최종 모델

```
pred_train = xgb_model.predict(X_train)
pred_test = xgb_model.predict(X_test)
train_score = roc_auc_score(y_train, pred_train)
test_score = roc_auc_score(y_test, pred_test)
test_score 

print('최종 훈련 정확도: ', xgb_model.score(X_train,y_train))
print('최종 테스트 정확도: ',xgb_model.score(X_test,y_test))

print('최종 모델 훈련 Roc_AUC_score:', train_score)
print('최종 모델 테스트 Roc_AUC_score:', test_score)
```
